```python
# Make it into a list of words, lowercase them
twtok_text = [
     [word for word in document.lower().split()]
      for document in o_content
 ]
twtok_text[3:10]

# Remove NaN's from the interviews
#o_inter = [x for x in o_interviews if str(x) != 'nan']

#twtok_inter = [
#     [word for word in document.split()]
#      for document in o_inter
#]
#twtok_inter[3]
```

```python
## For texts:
dictionary = corpora.Dictionary(twtok_text)
print(dictionary)
corpus = [dictionary.doc2bow(text) for text in twtok_text]
```

```python
## For textx:
total_topics = 5
lda = models.LdaModel(corpus, id2word=dictionary, num_topics=total_topics)
```

```python
pyLDAvis.enable_notebook()
panel = pyLDAvis.gensim.prepare(lda, corpus, dictionary, mds='TSNE')
panel
```
